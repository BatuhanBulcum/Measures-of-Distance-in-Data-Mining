# **Euclidean Distance**

*It can be simply explained as the ordinary distance between two points. It is one of the most used algorithms in the cluster analysis.   Mathematically it computes the root of squared differences between the coordinates between two objects.*

unnamed4.png

Ekran Alıntısı.PNG

```python
u = [] 
n = int(input("N= ")) 
for i in range(n): 
    new_element = int(input())
    u.append(new_element)    
print("U =",u)
```

```python
v = [] 
n = int(input("N= ")) 
for i in range(n): 
    new_element = int(input())
    v.append(new_element)    
print("V =",v)
```

```python
def Euclidean_Distance(u,v):
  distance=0
  for i,j in zip(u,v):
    distance += ((i-j)**2)
  return sqrt(distance)
print("U= ",u,"\nV= ",v,"\nEuclidean Distance(U,V)= ",Euclidean_Distance(u,v))
```

```python
def Dot_Product(u,v):
  Dot_Product=0
  for (i, j) in zip(u, v):
    Dot_Product += (i*j)
  return Dot_Product
print("Dot Product(U,V)= ",Dot_Product(u,v))
```

***Magnitude of a Vector***

Magnitude-formulas.png

```python
def Vector_Magnitude(m):
  Vector_Magnitude=0
  for i in m:
    Vector_Magnitude += (i**2)
  return sqrt(Vector_Magnitude)
print("||U||= ",Vector_Magnitude(u))
print("||V||= ",Vector_Magnitude(v))
```

***Cosine Similarity***

```python
def Cosine_Similarity(u,v):
  Cosine_Similarity = ((Dot_Product(u,v)) / Vector_Magnitude(u) * Vector_Magnitude(v))
  return Cosine_Similarity
print("U= ",u,"\nV= ",v,"\nCosine Similarity(U,V)= ",Cosine_Similarity(u,v))
```

# **Correlation**

*A correlation is a statistical measure of the relationship between two variables. The measure is best used in variables that demonstrate a linear relationship between each other. The fit of the data can be visually represented in a scatterplot. Using a scatterplot, we can generally assess the relationship between the variables and determine whether they are correlated or not.*

*The correlation coefficient is a value that indicates the strength of the relationship between variables. The coefficient can take any values from -1 to 1. The interpretations of the values are:*


*   *-1: Perfect negative correlation. The variables tend to move in opposite directions (i.e., when one variable increases, the other variable decreases).*
*   *0: No correlation. The variables do not have a relationship with each other.*
*   *1: Perfect positive correlation. The variables tend to move in the same direction (i.e., when one variable increases, the other variable also increases).*

***Standard Deviation***

*The standard deviation is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean of the set, while a high standard deviation indicates that the values are spread out over a wider range.*

220px-Normal-distribution-cumulative-density-function.svg.png

StandartSapmaFormulu.png



```python
def Standard_Deviatoin(u):
  mean = sum(u)/len(u)
  total = 0
  for i in u:
    total += ((mean-i) ** 2)
  return sqrt(total/(len(u)-1))
print("U= ",u,"\nStandard Deviatoin(U)= ",Standard_Deviatoin(u))
```

***Covariance***

*Covariance is a measure of the joint variability of two random variables. If the greater values of one variable mainly correspond with the greater values of the other variable, and the same holds for the lesser values, the covariance is positive. In the opposite case, when the greater values of one variable mainly correspond to the lesser values of the other, the covariance is negative.*

170px-Covariance_trends.svg.png

covariance-formula-surveygizmo-blog.png



```python
def Coveriance(u,v):
  mean_U = sum(u)/len(u)
  mean_V = sum(v)/len(v)
  total=0
  for i,j in zip(u,v):
    total+=(mean_U-i)*(mean_V-j)
  return total/(len(u)-1)
print("U= ",u,"\nV= ",v,"\nCoveriance(U,V)= ",Coveriance(u,v))
```

***Correlation***

```python
def Correlation(u,v): 
  return Coveriance(u,v)/(Standard_Deviatoin(u)*Standard_Deviatoin(v))
print("U= ",u,"\nV= ",v,"\nCorrelation(U,V)= ",Correlation(u,v))
```

# **Manhattan Distance**

*This determines the absolute difference among the pair of the coordinates.*

*Suppose we have two points P and Q to determine the distance between these points we simply have to calculate the perpendicular distance of the points from X-Axis and Y-Axis.* 

*In a plane with P at coordinate (x1, y1) and Q at (x2, y2)*

*Manhattan distance between P and Q = |x1 – x2| + |y1 – y2|*

6.PNG

0__9ljPf7RbVI5cVdG.png

*Here the total distance of the **Green** line gives the Manhattan distance between both the points.*

```python
def Manhattan_Distance(u,v):
  distance=0
  for i,j in zip(u,v):
    distance += abs(i-j)
  return distance
print("U= ",u,"\nV= ",v,"\nManhattan Distance(U,V)= ",Manhattan_Distance(u,v)) 
```

# **Minkowski Distance**


*It is the generalized form of the Euclidean and Manhattan Distance Measure. In an N-dimensional space, a point is represented as,*

**(x1, x2, ..., xN)**

Consider two points P1 and P2:

**P1: (X1, X2, ..., XN)**    
**P2: (Y1, Y2, ..., YN)**

*Then, the Minkowski distance between P1 and P2 is given as:*

*   *When p = 2, Minkowski distance is same as the Euclidean distance.*
*   *When p = 1, Minkowski distance is same as the Manhattan distance.*

```python
p = int(input("P= "))
def Minkowski_Distance(u,v,p): 
  distance=0
  for i,j in zip(u,v):
    distance += (abs(i-j)**p)
  return distance**(1/p)
print("U= ",u,"\nV= ",v,"\nMinkowski_Distance(U,V)= ",Minkowski_Distance(u,v,p))
```

# **Hamming Distance**

*The Hamming distance between two strings of equal length is the number of positions at which the corresponding symbols are different. In other words, it measures the minimum number of substitutions required to change one string into the other, or the minimum number of errors that could have transformed one string into the other.*
